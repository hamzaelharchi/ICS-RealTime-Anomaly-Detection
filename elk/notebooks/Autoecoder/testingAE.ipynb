```python
import onnxruntime as ort
import joblib
import numpy as np
import pandas as pd
import json
```

```python
# Load ONNX LSTM Autoencoder model
myAutoEncoder = ort.InferenceSession("../model.onnx")
input_name = myAutoEncoder.get_inputs()[0].name  # Get input tensor name
output_name = myAutoEncoder.get_outputs()[0].name  # Get output tensor name
print('Model loaded')
```

```python

# Generate random DataFrame with 300 rows and 69 features
data = np.random.rand(10, 69).astype(np.float32)  # Random float values between 0 and 1
df = pd.DataFrame(data, columns=[f"feature_{i}" for i in range(69)])

print("Generated DataFrame:")
print(df.head())  # Display first 5 rows of the DataFrame

# Convert DataFrame to NumPy array and reshape for ONNX model
input_data = df.to_numpy().reshape(1, 10, 69)  # Shape (1, 300, 69)

# Run the ONNX model
reconstructed = myAutoEncoder.run([output_name], {input_name: input_data})

# Output the result
print("Reconstructed Output:")
print(np.array(reconstructed).shape)
print(np.array(reconstructed))
```
